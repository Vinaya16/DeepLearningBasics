```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris, load_digits
import matplotlib.pyplot as plt
```

```python
import matplotlib.colors as mcolors
```

```python
colors = list(mcolors.TABLEAU_COLORS)
colors
```

```python
def tsne_and_plot(load_data):
    X = load_data()['data']
    Y = load_data()['target']
    embedded_X = TSNE(n_components=2).fit_transform(X)
    plt.figure(figsize=(7, 5))
    id2class = load_data()['target_names']
    for label, color in zip(set(Y), colors[:len(id2class)]):
        plt.scatter(embedded_X[Y == label,0], embedded_X[Y == label,1], c = color, label=id2class[label])
    plt.legend()
```

```python
tsne_and_plot(load_iris)
```

```python
tsne_and_plot(load_iris)
```

```python
tsne_and_plot(load_digits)
```

```python
tsne_and_plot(load_digits)
```
