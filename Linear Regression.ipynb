## Linear Regression using PyTorch

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from IPython import display
import time
```

### Generate data

```python
np.random.seed(11)
N = 40
x = np.random.uniform(0, 1, N)
y = 0.5 + 4*x + np.random.normal(0, 0.25, N)
plt.scatter(x, y)
plt.show()
```

### Model: 1-Neuron Neural Network

![title](img/1neuronmodel.png)

### Converting x & y to Tensor

```python
X = torch.ones(2, 40)
X[1,] = torch.from_numpy(x)
X[:,:5]
```

```python
Y = torch.from_numpy(y).unsqueeze(0)
Y[:,:5]
```

### Define loss function - MSE

```python
def mse_loss():
    return ((torch.mm(w, X) - Y) ** 2).mean()
```

### Define learning rate and initial random weights

```python
alpha = 0.7 # learning rate
w = torch.tensor([[2.5, -1]]) # initial weights
```

### Training

```python
epochs = 50

for _ in range(epochs):
    w.requires_grad_(True)
    loss = mse_loss()
    print('MSE=', loss.item())
    loss.backward()
    w.requires_grad_(False)
    
    # update weights
    # w.grad: partial derivative of loss w.r.t. w0 and w1
    w = w - alpha * w.grad 
    
    # see how approximated lines are changing
    u = np.linspace(0, 1, 100)
    v = w[0,0].item() + u * w[0,1].item()
    plt.scatter(x, y)
    plt.plot(u, v)
    plt.show()
```
